```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
df.head(10)
```

```python
#how many pokemon exists with an Attack value greater than 150

df.loc[df['Attack'] > 150]
```

```python
df.loc[df['Attack'] > 150].shape
```

```python
#query method to achieve the same
df.query("Attack > 150")
```

```python
#select all the pokemons with a speed of 10 or less
df.query("Speed <= 10")
```

```python
#how many pokemons have a Sp.Def value of 25 or less?
df.loc[df['Sp. Def']<=25]
```

```python
#select all the legendary pokemons
legendary_df = df.loc[df['Legendary'] == True]
print(legendary_df.head(10))
```

```python
#total number of legendary pokemons

df['Legendary'].sum()
```

```python
#find the pokemon which is a outlier 
#outlier : doesnot follow the general trend of a dataset, can be the one which has a lot of defence but less attack

df.sort_values(by=['Defense', 'Attack'], ascending=[False, True]).head(1)
```

```python
df.sort_values(by=['Defense', 'Attack'], ascending=[True, False]).head(1)
```

```python
#how many fire-flying pokemons are there?
# type1 = fire , type2 = flying

((df['Type 1'] == 'Fire') & (df['Type 2'] == 'Flying')).sum()
```

```python
df.loc[(df['Type 1'] == 'Fire') & (df['Type 2'] == 'Flying') ]
```

```python
#using the query method to get the fire-flying pokemons

df.query("`Type 1` == 'Fire' and `Type 2` == 'Flying'")
```

```python
#how many poison pokemons are across both types
#it means it is asking to find out the number of pokemons which are poison in Type 1 or Type 2

((df['Type 1'] == 'Poison') | (df['Type 2'] == 'Poison')).sum()
```

```python
df.loc[((df['Type 1'] == 'Poison') | (df['Type 2'] == 'Poison'))]
```

```python
#using query method to achieve the same

df.query(" `Type 1` == 'Poison' or `Type 2` == 'Poison'")
```
